Сумма длин окружностей переднего и заднего колёс экипажа равна 5 м., переднее колесо на расстоянии 64 м.
сделало столько же оборотов, сколько заднее колесо на расстоянии 96 м. Определить длину окружности
каждого колеса.

```python
from sympy.abc import x
from sympy import Eq
from sympy import solve

sol = solve([Eq(64 / x, 96/(5-x))], [x])
n1 = sol[x]
n2 = 5 - n1
print(sol)
print("Length:",[str(n1) + " m", str(n2) + " m"])
```

```python
from sympy.abc import x,y
from sympy import Eq
from sympy import solve

sol = solve([Eq(x + y, 5), Eq(64 / x, 96 / y)], [x,y])
n1 = sol[x]
n2 = sol[y]
print(sol)
print("Length:",[str(n1) + " m", str(n2) + " m"])
```
